# **Cálculo del coeficiente de correlación (\( r \))**

El coeficiente de correlación (\( r \)) mide la relación entre dos variables, \( X \) e \( Y \). Su valor varía entre \(-1\) y \(1\), donde:

- \( r = 1 \): Relación positiva perfecta.
- \( r = -1 \): Relación negativa perfecta.
- \( r = 0 \): No hay relación entre las variables.

La fórmula es:

$$
r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \cdot \sum{(Y_i - \bar{Y})^2}}}
$$

### **Pasos:**
1. Calcula las medias:
   $$
   \bar{X} = \frac{\sum{X_i}}{n}, \quad \bar{Y} = \frac{\sum{Y_i}}{n}
   $$

2. Resta las medias de cada valor:
   $$
   (X_i - \bar{X}), \quad (Y_i - \bar{Y})
   $$

3. Multiplica las desviaciones correspondientes:
   $$
   (X_i - \bar{X})(Y_i - \bar{Y})
   $$

4. Suma los resultados:
   $$
   \sum{(X_i - \bar{X})(Y_i - \bar{Y})}
   $$

5. Calcula el denominador:
   $$
   \sqrt{\sum{(X_i - \bar{X})^2} \cdot \sum{(Y_i - \bar{Y})^2}}
   $$

6. Divide para obtener \( r \):
   $$
   r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \cdot \sum{(Y_i - \bar{Y})^2}}}
   $$


# Resultado Final

### Gráfico superior izquierdo:
\( r = -0.17 \)

### Gráfico superior derecho:
\( r = 1 \)

### Gráfico inferior izquierdo:
\( r = -0.96 \)

### Gráfico inferior derecho:
\( r = 0.505 \)

<span style="color: #153CEE;">[8] Solo para practicar1</span>. En México, 6% de los 30 millones de usuarios de internet que hay utilizan Twitter. (Torres, 2010). Si conoces a 15 personas que sean asiduas a las redes sociales:  
a) ¿Cuál es la probabilidad de que 9 o más utilicen Twitter?  

| **X** | **Y** |
|-------|-------|
| 7     | 2     |
| 8     | 1     |
| 9     | 2     |
| 11    | 1     |
| 6     | 3     |


```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Datos de la tabla
X = np.array([7, 8, 9, 11, 6])
Y = np.array([2, 1, 2, 1, 3])

# a) Traza el diagrama de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(X, Y, color='blue', label='Datos', alpha=0.7)
plt.title("Diagrama de Dispersión", fontsize=14)
plt.xlabel("X", fontsize=12)
plt.ylabel("Y", fontsize=12)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# b) Cálculo de coeficientes de correlación y determinación
correlacion, _ = np.corrcoef(X, Y)[0, 1], linregress(X, Y).pvalue
r_cuadrado = correlacion ** 2

print(f"Coeficiente de correlación (r): {correlacion:.4f}")
print(f"Coeficiente de determinación (r^2): {r_cuadrado:.4f}")

# Interpretación
if correlacion > 0:
    interpretacion = "Relación positiva"
elif correlacion < 0:
    interpretacion = "Relación negativa"
else:
    interpretacion = "Sin relación"
print(f"Interpretación del coeficiente de correlación: {interpretacion}")

# c) Ecuación de regresión
pendiente, ordenada, _, _, _ = linregress(X, Y)

print(f"Ecuación de regresión: Y = {pendiente:.4f}X + {ordenada:.4f}")
print(f"Interpretación de la pendiente: Por cada unidad que aumenta X, Y cambia en promedio {pendiente:.4f} unidades.")
print(f"Interpretación de la ordenada al origen: Cuando X = 0, el valor promedio estimado de Y es {ordenada:.4f}.")
```

<span style="color: #153CEE;">[10] Negocios</span>. Una empresa desea desarrollar un modelo para pronosticar sus ventas conbase en los gastos que realiza en publicidad. La siguiente es la información con la que cuenta:

| Ventas (Millones de pesos) | Gastos de Publicidad (Millones de pesos) |
|-----------------------------|------------------------------------------|
| 2                           | 4                                        |
| 3                           | 5                                        |
| 4                           | 4                                        |
| 5                           | 6                                        |
| 7                           | 7                                        |
| 6                           | 6                                        |
| 5                           | 7                                        |
| 8                           | 9                                        |
| 9                           | 11                                       |

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Datos de la tabla
ventas = np.array([2, 3, 4, 5, 7, 6, 5, 8, 9])  # Ventas (Millones de pesos)
gastos_publicidad = np.array([4, 5, 4, 6, 7, 6, 7, 9, 11])  # Gastos de publicidad (Millones de pesos)

# a) Traza el diagrama de dispersión
plt.figure(figsize=(8, 6))
plt.scatter(gastos_publicidad, ventas, color='blue', label='Datos', alpha=0.7)
plt.title("Diagrama de Dispersión: Ventas vs Gastos de Publicidad", fontsize=14)
plt.xlabel("Gastos de Publicidad (Millones de pesos)", fontsize=12)
plt.ylabel("Ventas (Millones de pesos)", fontsize=12)
plt.grid(alpha=0.3)
plt.legend()
plt.show()

# b) Cálculo de coeficientes de correlación y determinación
correlacion = np.corrcoef(gastos_publicidad, ventas)[0, 1]
r_cuadrado = correlacion ** 2

print(f"Coeficiente de correlación (r): {correlacion:.4f}")
print(f"Coeficiente de determinación (r^2): {r_cuadrado:.4f}")

# Interpretación
if correlacion > 0:
    interpretacion = "Relación positiva: a medida que aumentan los gastos de publicidad, aumentan las ventas."
elif correlacion < 0:
    interpretacion = "Relación negativa: a medida que aumentan los gastos de publicidad, disminuyen las ventas."
else:
    interpretacion = "No hay relación entre los gastos de publicidad y las ventas."
print(f"Interpretación del coeficiente de correlación: {interpretacion}")

# c) Determinación de la ecuación de regresión
pendiente, ordenada, _, _, _ = linregress(gastos_publicidad, ventas)

print(f"Ecuación de regresión: Ventas = {pendiente:.4f} * Gastos_Publicidad + {ordenada:.4f}")
print(f"Interpretación de la pendiente: Por cada millón de pesos adicional gastado en publicidad, las ventas aumentan en promedio {pendiente:.4f} millones de pesos.")
print(f"Interpretación de la ordenada al origen: Cuando no se gasta en publicidad, se estiman ventas promedio de {ordenada:.4f} millones de pesos.")
```

b) Calcula los coeficientes de correlación y de determinación. Interpreta.

```python
import pandas as pd

# Datos
data = {
    'País': ['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Costa Rica', 
             'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'México', 
             'Nicaragua', 'Panamá', 'Paraguay', 'Perú', 'Uruguay', 'Venezuela'],
    'Promedio educativo (años)': [9.3, 9.2, 7.2, 9.7, 7.4, 8.3, 7.6, 7.7, 
                      4.1, 6.5, 8.7, 5.7, 9.4, 7.8, 9.6, 8.4, 6.2],
    'PIB per cápita (2010)': [14931, 4502, 10847, 14780, 8959, 11143, 8170, 6660, 4761, 3845, 14192, 2632, 13210, 4629, 9016,
                  14022, 11820]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular coeficiente de correlación
correlation = df['Promedio educativo (años)'].corr(df['PIB per cápita (2010)'])
print(f'Coeficiente de correlación: {correlation:.4f}')

# Calcular coeficiente de determinación (R²)
determination = correlation ** 2
print(f'Coeficiente de determinación (R²): {determination:.4f}')
```

c) Determina la ecuación de regresión. Interpreta los coeficientes de regresión.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos
promedio_educativo = [9.3, 9.2, 7.2, 9.7, 7.4, 8.3, 7.6, 7.7, 
                      4.1, 6.5, 8.7, 5.7, 9.4, 7.8, 9.6, 8.4, 6.2]
pib_per_capita = [14931, 4502, 10847, 14780, 8959, 11143, 8170, 6660, 4761, 3845, 14192, 2632, 13210, 4629, 9016,
                  14022, 11820]

# Convertir los datos a formato numpy para la regresión
X = np.array(promedio_educativo).reshape(-1, 1)  # Promedio educativo (variable independiente)
y = np.array(pib_per_capita)  # PIB per cápita (variable dependiente)

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener los coeficientes de regresión
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

# Mostrar la ecuación de regresión
print(f"Ecuación de regresión: PIB per cápita = {intercepto:.2f} + {pendiente:.2f} * Promedio educativo")

# Predecir valores para graficar la línea de regresión
x_pred = np.linspace(min(promedio_educativo), max(promedio_educativo), 100).reshape(-1, 1)
y_pred = modelo.predict(x_pred)

# Graficar los datos y la línea de regresión
plt.figure(figsize=(12, 8))
plt.scatter(promedio_educativo, pib_per_capita, color='green', alpha=0.7, edgecolor='black', label="Datos")
plt.plot(x_pred, y_pred, color='red', label="Línea de regresión")

# Etiquetas del gráfico
plt.title("Regresión lineal: Promedio Educativo vs PIB per Cápita", fontsize=16)
plt.xlabel("Promedio Educativo (Años)", fontsize=14)
plt.ylabel("PIB per Cápita (USD)", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

b) Calcula los coeficientes de correlación y de determinación. Interprétalos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de la tabla
data = {
    'País': ['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Costa Rica', 
             'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'México', 
             'Nicaragua', 'Panamá', 'Paraguay', 'Perú', 'Uruguay', 'Venezuela'],
    'Confianza en los partidos políticos': [14.6, 20.2, 20.4, 21.8, 23.5, 23.9, 
                                            14.4, 38.3, 21.6, 16.2, 18.7, 
                                            14.3, 31.2, 31.2, 11.2, 46.1, 33.8],
    'Percepción de la democracia': [5.8, 5.5, 6.0, 6.0, 6.1, 6.5, 6.0, 
                                    5.4, 5.5, 5.5, 5.0, 5.4, 5.9, 
                                    5.0, 5.0, 7.1, 6.6]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Calcular el coeficiente de correlación
correlacion = df['Confianza en los partidos políticos'].corr(df['Percepción de la democracia'])
print(f'Coeficiente de correlación: {correlacion:.4f}')

# Calcular el coeficiente de determinación (R²)
r_cuadrado = correlacion ** 2
print(f'Coeficiente de determinación (R²): {r_cuadrado:.4f}')

# Interpretación
if correlacion > 0:
    interpretacion = "Hay una relación positiva entre la confianza en los partidos políticos y la percepción de la democracia."
elif correlacion < 0:
    interpretacion = "Hay una relación negativa entre la confianza en los partidos políticos y la percepción de la democracia."
else:
    interpretacion = "No hay una relación lineal entre la confianza en los partidos políticos y la percepción de la democracia."

print(interpretacion)
```

c) Determina la ecuación de regresión. Interpreta los coeficientes de regresión.

```python
import pandas as pd
import statsmodels.api as sm

# Datos de la tabla
data = {
    'País': ['Argentina', 'Bolivia', 'Brasil', 'Chile', 'Colombia', 'Costa Rica', 
             'Ecuador', 'El Salvador', 'Guatemala', 'Honduras', 'México', 
             'Nicaragua', 'Panamá', 'Paraguay', 'Perú', 'Uruguay', 'Venezuela'],
    'Confianza en los partidos políticos': [14.6, 20.2, 20.4, 21.8, 23.5, 23.9, 
                                            14.4, 38.3, 21.6, 16.2, 18.7, 
                                            14.3, 31.2, 31.2, 11.2, 46.1, 33.8],
    'Percepción de la democracia': [5.8, 5.5, 6.0, 6.0, 6.1, 6.5, 6.0, 
                                    5.4, 5.5, 5.5, 5.0, 5.4, 5.9, 
                                    5.0, 5.0, 7.1, 6.6]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Definir variables independientes y dependientes
X = df['Confianza en los partidos políticos']
y = df['Percepción de la democracia']

# Añadir una constante a la variable independiente
X = sm.add_constant(X)

# Ajustar el modelo de regresión
modelo = sm.OLS(y, X).fit()

# Resumen del modelo
print(modelo.summary())

# Obtener la ecuación de regresión
intercepto = modelo.params[0]
coeficiente_confianza = modelo.params[1]

print(f'Ecuación de regresión: y = {intercepto:.4f} + {coeficiente_confianza:.4f} * (Confianza en los partidos políticos)')
```

# Análisis del Desempeño Escolar

## a) Traza el diagrama de dispersión
Para crear un diagrama de dispersión, se grafican los datos de "Grado Total en Educación" en el eje X y "Grado Promedio de Escolaridad" en el eje Y. Puedes usar herramientas como Excel, Google Sheets o software de análisis de datos como Python (con Matplotlib o Seaborn).

## b) Coeficientes de correlación y determinación
Los coeficientes se pueden calcular utilizando las siguientes fórmulas:

### Coeficiente de correlación (r)
$$
r = \frac{n(\Sigma xy) - (\Sigma x)(\Sigma y)}{\sqrt{[n\Sigma x^2 - (\Sigma x)^2][n\Sigma y^2 - (\Sigma y)^2]}}
$$

### Coeficiente de determinación (r²)
$$
r^2 = r^2
$$

Donde:
- \(n\) es el número de observaciones (32).
- \(x\) es el "Grado Total en Educación".
- \(y\) es el "Grado Promedio de Escolaridad".

## c) Ecuación de regresión
La ecuación de regresión lineal se expresa como:
$$
y = mx + b
$$

Donde:
- \(y\) es el grado promedio de escolaridad.
- \(x\) es el grado total en educación.
- \(m\) es la pendiente, que se calcula como:
$$
m = \frac{n(\Sigma xy) - (\Sigma x)(\Sigma y)}{n(\Sigma x^2) - (\Sigma x)^2}
$$
- \(b\) es la intersección con el eje Y, que se calcula como:
$$
b = \frac{\Sigma y - m(\Sigma x)}{n}
$$

## d) Estimación del grado de escolaridad
Si se otorgan 12,000 millones de pesos, se puede usar la ecuación de regresión para estimar el grado de escolaridad:
$$
\hat{y} = m(12000) + b
$$

## e) Debate sobre la optimización de recursos
Conforme a los resultados analizados, se pueden considerar varias estrategias para optimizar el uso de recursos en educación:
- *Inversión focalizada*: Dirigir fondos hacia áreas con mayor necesidad, como infraestructura y capacitación docente.
- *Programas de evaluación*: Implementar evaluaciones regulares para medir el impacto de las inversiones.
- *Participación comunitaria*: Involucrar a la comunidad en el proceso educativo para fomentar un entorno de aprendizaje positivo.
- *Uso de tecnología*: Integrar herramientas tecnológicas para mejorar la enseñanza y el aprendizaje.

## Resultados Ficticios para Ejemplo
Supongamos que después de calcular, obtenemos:
- \(r = 0.75\) (correlación positiva moderada).
- \(r^2 = 0.56\) (56% de la variación en el grado de escolaridad se explica por el grado total en educación).
- Ecuación de regresión: \(y = 0.0005x + 6.5\).

Si se sustituye 12,000 millones en la ecuación:
$$
\hat{y} = 0.0005(12000) + 6.5 = 6.5 + 6 = 12.5
$$

Esto indicaría que el grado promedio de escolaridad estimado sería de 12.5.

## Conclusión
Estos análisis brindan una visión cuantitativa del impacto de la inversión educativa. El uso eficiente de los recursos es fundamental para mejorar los resultados en evaluaciones educativas y para garantizar una educación de calidad.

```python
import pandas as pd
import statsmodels.api as sm

# Datos de la tabla
data = {
    'Año': [1950, 1960, 1970, 1980, 1990, 2000, 2010],
    'Población total (miles de habitantes)': [25791, 34923, 48225, 68776, 81250, 97483, 112337]
}

# Crear DataFrame
df = pd.DataFrame(data)

# Definir variables independientes y dependientes
X = df['Año']
y = df['Población total (miles de habitantes)']

# Añadir una constante a la variable independiente
X = sm.add_constant(X)

# Ajustar el modelo de regresión
modelo = sm.OLS(y, X).fit()

# Resumen del modelo
print(modelo.summary())

# Obtener la ecuación de regresión
intercepto = modelo.params[0]
coeficiente_año = modelo.params[1]

print(f'Ecuación de regresión: y = {intercepto:.2f} + {coeficiente_año:.2f} * Año')
```

d) Determina la ecuación de regresión e interpreta los coeficientes.

La ecuación de regresión lineal se expresa generalmente como:

$$
Y = a + bX
$$

donde:
- \(Y\) es la variable dependiente (consumo).
- \(X\) es la variable independiente (ingreso).
- \(a\) es el intercepto.
- \(b\) es la pendiente.

Para calcular los coeficientes \(a\) y \(b\), se utilizan las siguientes fórmulas:

1. **Pendiente (\(b\))**:

$$
b = \frac{n(\sum XY) - (\sum X)(\sum Y)}{n(\sum X^2) - (\sum X)^2}
$$

2. **Intercepto (\(a\))**:

$$
a = \frac{\sum Y - b(\sum X)}{n}
$$

donde:
- \(n\) es el número de observaciones.
- \(\sum XY\) es la suma del producto de \(X\) y \(Y\).
- \(\sum X\) es la suma de los ingresos.
- \(\sum Y\) es la suma de los consumos.
- \(\sum X^2\) es la suma de los cuadrados de los ingresos.

Ejemplo de resultado:

$$
Y = 300 + 0.5X
$$
